### [Step1] 데이터 로드

```python
import pandas as pd

df = pd.read_csv("datasets/학생.csv")
df
```

### [Step2] 데이터 파악하기

```python
df.info()
```

```python
df.describe()
```

### [Step3] 데이터 라벨(학교) 만들기

```python
school = []

# 학교의 종류를 나타내는 열을 만들어줍니다.
for idx, row in df.iterrows():
    if row.학교명.find('고등학교') > 0: # 학교이름에 고등학교가 들어가는 지 확인합니다.
        school.append('고등학교')
    elif row.학교명.find('중학교') > 0: # 학교이름에 중학교가 들어가는 지 확인합니다.
        school.append('중학교')
    elif row.학교명.find('초등학교') > 0: # 학교이름에 초등학교가 들어가는 지 확인합니다.
        school.append('초등학교')
    else:
        school.append(None)

df['학교'] = school
df
```

### [Step4] 데이터 형태 변형하기

```python
df['성별'] = df.apply(lambda x : 0 if x.성별 == '남' else 1, axis = 1) # 성별을 숫자로 바꿔줍니다( 남자 = 0, 여자 = 1)
df['학교'] = df.apply(lambda x : 0 if x.학교 == '초등학교' else (1 if x.학교 == '중학교' else (2 if x.학교 == '고등학교' else None)), axis = 1) # 학교를 숫자로 바꿔줍니다.( 초등 = 0, 중등 = 1, 고등 = 2)

df
```

```python
df = df[['성별', '키', '몸무게','학교']]
df
```

### [Step5] 데이터 시각화하기

```python
import matplotlib.pyplot as plt

gender = df.value_counts('성별') # 성별별로 데이터의 수를 셉니다.
plt.pie(gender) # 원 그래프를 그립니다.
plt.legend(['남자','여자'])
plt.title('성별 비율')
plt.show()
```

```python
# 남학생
df_male = df[df['성별'] == 0]

df0_male = df_male[df_male['학교'] == 0] # 초등학교에 다니는 남자 학생들
df1_male = df_male[df_male['학교'] == 1] # 중학교에 다니는 남자 학생들
df2_male = df_male[df_male['학교'] == 2] # 고등학교에 다니는 남자 학생들

fig = plt.figure(figsize = (12,9))

plt.scatter(df0_male['키'], df0_male['몸무게'], c = 'b', s = 10, alpha = 0.3) # 초등학교에 다니는 남자 학생들을 파란색으로 설정
plt.scatter(df1_male['키'], df1_male['몸무게'], c = 'g', s = 10, alpha = 0.3) # 중학교에 다니는 남자 학생들을 초록색으로 설정
plt.scatter(df2_male['키'], df2_male['몸무게'], c = 'r', s = 10, alpha = 0.3) # 고등학교에 다니는 남자 학생들 빨간색으로 설정
plt.legend(['초등학생','중학생','고등학생'])
plt.title('학생 키 몸무게 분포(남자)')
plt.xlabel('키(cm)')
plt.ylabel('몸무게(kg)')

# 여학생
df_female = df[df['성별'] == 1]

df0_female = df_female[df_female['학교'] == 0] # 초등학교에 다니는 여자 학생들
df1_female = df_female[df_female['학교'] == 1] # 중학교에 다니는 여자 학생들
df2_female = df_female[df_female['학교'] == 2] # 고등학교에 다니는 여자 학생들

fig = plt.figure(figsize = (12,9))

plt.scatter(df0_female['키'], df0_female['몸무게'], c = 'b', s = 10, alpha = 0.3) # 초등학교에 다니는 여자 학생들을 파란색으로 설정
plt.scatter(df1_female['키'], df1_female['몸무게'], c = 'g', s = 10, alpha = 0.3) # 중학교에 다니는 여자 학생들을 초록색으로 설정
plt.scatter(df2_female['키'], df2_female['몸무게'], c = 'r', s = 10, alpha = 0.3) # 고등학교에 다니는 여자 학생들을 빨간색으로 설정
plt.legend(['초등학생','중학생','고등학생'])
plt.title('학생 키 몸무게 분포(여자)')
plt.xlabel('키(cm)')
plt.ylabel('몸무게(kg)')

plt.show()
```

### [Step6] 데이터 넘파이로 변형하기

```python
import numpy as np

x = np.array(df.iloc[:,:-1]) # input data를 numpy로 변형
y = np.array(df.iloc[:,-1]) # label data를 numpy로 변형
```

### [Step7] 모델 구성 및 학습하기

```python
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(random_state=1, max_iter=20,hidden_layer_sizes=(128,),verbose = True).fit(x,y) # 모델을 학습시킵니다.
```

### [Step8] 어떤 학생인지 예측하는 프로그램 만들기 

```python
gender = input('성별을 입력하세요(남/여)')
if gender == '남':
    gender = 0
elif gender == '여':
    gender = 1
else:
    print('잘못된 성별입니다.')
    
height = int(input('키를 입력하세요(cm)'))
weight = int(input('몸무게를 입력하세요(kg)'))

transform = {0:'초등학생', 1:'중학생', 2:'고등학생'}

print(f'당신은 {transform[clf.predict([[gender,height,weight]])[0]]} 입니다!!') 
```
